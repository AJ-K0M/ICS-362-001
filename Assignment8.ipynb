```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("auto-mpg.data", delim_whitespace=True, header=None)
print(df.head())
```

```python
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'car_name']
print(df.head())
```

```python
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['horsepower'] = df['horsepower'].astype(float)
data=df.dropna(subset=['horsepower'])
print(df.head())
```

```python
df['origin'] = df['origin'].replace({1: 'USA', 2: 'Asia', 3: 'Europe'})
print(df.head())
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='cylinders', hue='origin', data=df)
plt.title('Distribution of Cylinders by Origin')
plt.xlabel('Number of Cylinders')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.lmplot(x='horsepower', y='weight', hue='origin', data=df)
plt.title('Horsepower and Weight by Origin')
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.lmplot(x='acceleration', y='mpg', hue='origin', data=df)
plt.title('Acceleration and MPG by Origin')
plt.xlabel('Acceleration')
plt.ylabel('MPG')
plt.show()
```

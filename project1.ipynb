```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "airline_delays.csv"
df = pd.read_csv(file_path)

df["Airline"] = df["Airline"].ffill()

df_delayed = df[df["Status"] == "delayed"].set_index("Airline").drop(columns=["Status"])

total_delays = df_delayed.sum(axis=1)
avg_delays = df_delayed.mean(axis=1)

plt.figure(figsize=(8, 5))
plt.bar(total_delays.index, total_delays.values)
plt.title("Total Delayed Flights")
plt.xlabel("Airline")
plt.ylabel("Number of Delays")
plt.show()

df_delayed.T.plot(kind="bar", figsize=(8, 5))
plt.title("Average Delays by Destination")
plt.xlabel("Destination")
plt.ylabel("Number of Delays")
plt.show()

print("\nThe dataset shows flight delays for two airlines: Alaska Airlines and AM West Airlines.\nEach airline has flights to multiple destinations, and I analyze how frequently these flights experience delays.")
print("\nTotal Delays per Airline:")
print(total_delays)
print("\nAverage Delays per Airline:")
print(avg_delays)
print("\nAlaska Airlines has fewer delays than AM West Airlines,\nThis is because AM West has more flights, especially to Phoenix, \nleading to a higher number of delays overall.")


```

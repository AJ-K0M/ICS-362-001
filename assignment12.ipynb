```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

columns = ['class', 'cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
    'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
    'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
    'stalk-surface-below-ring', 'stalk-color-above-ring',
    'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
    'ring-type', 'spore-print-color', 'population', 'habitat']

df = pd.read_csv('agaricus-lepiota.data', names=columns)
df.head()
```

```python
df = df[['class', 'odor', 'population']]
df.columns = ['edible_or_poisonous', 'odor', 'population']
df.head()
```

```python
df['edible_or_poisonous'] = df['edible_or_poisonous'].replace({'e': 0, 'p': 1})
df['odor'] = df['odor'].astype('category').cat.codes
df['population'] = df['population'].astype('category').cat.codes
df.head()
```

```python
sns.set(style="whitegrid")
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

sns.countplot(x='edible_or_poisonous', data=df, ax=axs[0])
axs[0].set_title('Edible vs Poisonous')
axs[0].set_xlabel('0 = Edible, 1 = Poisonous')

sns.countplot(x='odor', data=df, ax=axs[1])
axs[1].set_title('Odor Distribution')

sns.countplot(x='population', data=df, ax=axs[2])
axs[2].set_title('Population Distribution')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
sns.scatterplot(x='odor', y='edible_or_poisonous', data=df)
plt.title('Edible/Poisonous vs Odor')
plt.xlabel('Odor')
plt.ylabel('Edible (0) or Poisonous (1)')

plt.subplot(1, 2, 2)
sns.scatterplot(x='population', y='edible_or_poisonous', data=df)
plt.title('Edible/Poisonous vs Population')
plt.xlabel('Population')
plt.ylabel('Edible (0) or Poisonous (1)')

plt.tight_layout()
plt.show()
```
